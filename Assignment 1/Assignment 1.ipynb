```python
sns.histplot(data=data, x="Weight", color="green", kde=True,label="weight")
plt.title("Weight distribution")
plt.xlabel("Weight(Kg)")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

```python
sns.scatterplot(x="Age",y="Height", hue="Weight",data=data)
```

```python
sns.jointplot(x="Height", y="Weight", data=data,kind="reg", color="green")
```

```python
sns.barplot(x="CALC",y="Age", data=data, hue="Gender",palette="Set1")
plt.title("Consumption of alcohol vs Age")
plt.show()
```

```python
sns.countplot(data=data, x="CALC",palette="Set1")
plt.title("distribution of CALC values")
plt.xlabel("CALC")
plt.ylabel("Counts")
plt.show()
```
